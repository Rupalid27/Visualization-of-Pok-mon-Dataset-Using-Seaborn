```python
#pre-format DataFrame
stats_df = pokemon_df.drop(["#",'Total','Stage', 'Legendary'], axis = 1)

# New boxplot using stats_df
sns.boxplot(data = stats_df)
plt.show()
```

```python
# set theme
sns.set_style("darkgrid")

# Violin plot
plt.figure(figsize = (10,5))
sns.violinplot( x = "Type 1", y = "Attack", data = pokemon_df)
plt.show()
```

```python
# Swarn plot with Pokemon color palette
plt.figure(figsize = (10,4))
sns.swarmplot(x = "Type 1", y = "Attack", data =  pokemon_df)
plt.show()
```

```python
# set figure size

plt.figure(figsize=(10,6))

# Create Plot
# Plot 1
sns.violinplot( x = "Type 1", 
               y = "Attack", 
               data = pokemon_df, 
               inner = None) # Inner None -> Remove the bars inside the Violins

#Plot 2 
sns.swarmplot(x = "Type 1", y = "Attack", 
              data =  pokemon_df, 
              color = "k", # Make points Black
              alpha = 0.3) # and slightle transparent

# set title with Matplotlib
plt.title("Attack by Type") 
plt.show()
```

```python
# heatmap helps you visualise matrix like data
# calculated Correlation
corr = stats_df.corr()
#print(corr)

#heatmap
sns.heatmap(corr, cmap = "YlGnBu", annot = True) # Annot -> Label
plt.show()
    
```

```python
plt.figure(figsize =  (10,5))

#count plot (a.k.a  Bar plot)
sns.countplot(x = "Type 1", data = pokemon_df)
# sns.countplot(y = "Type 1", data = pokemon_df)

#rotate x-label
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize =  (10,5))

#count plot (a.k.a  Bar plot)
#sns.countplot(x = "Type 1", data = pokemon_df)
sns.countplot(y = "Type 1", data = pokemon_df)

#rotate x-label
plt.xticks(rotation = 1)
plt.show()
```

```python
sns.countplot(x = "Legendary", data = pokemon_df)
plt.show()
```

```python
# Pointplot
sns.pointplot(data = pokemon_df, x = "Stage", y = "Attack")
plt.show()
```

```python
## Kernel density estimation Plot

sns.displot(x = "Attack", y = "Defense",kind = "kde", data = pokemon_df, height = 7) # Kind = Type of plot
plt.show()
```
